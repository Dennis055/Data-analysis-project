```python
from FinMind.Data import Load
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import seaborn as sns
```

```python
stock_list  = []
for i in range(2000,3000):
    stock_name = str(i) + '.TW'
    stock_list.append(stock_name)
    stock_name = ''
```

```python
stock1 = Load.FinData(dataset = 'FinancialStatements',select = ['1101.TW'],date = '2018-9-10')
for stock_n in stock_list:
    stock = Load.FinData(dataset = 'FinancialStatements',select = [stock_n],date = '2018-9-10')
    stock1 = stock1.append(stock)
stock1.head()
```

```python
stock1.index = stock1.stock_id
data = stock1.ix[:,:-3]
data.head()
```

```python
data.corr()
```

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize = (20,15))
sns.heatmap(data.corr())
plt.xticks(rotation = 25)
```

```python
target_feature = data.corr()['Revenue']
relative_feature = target_feature[target_feature>0.5]
print(relative_feature)
```

```python
from sklearn import metrics , cluster
```

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
data_scaled = scale(data)
n_components = 5
pca = PCA(n_components= 5 , whiten=True).fit(data_scaled)
total_explain = 0
ex = []
print(pca.explained_variance_ratio_)
for n in range(n_components):
    total_explain += pca.explained_variance_ratio_[n]
    ex.append(total_explain)
pd.DataFrame([ex,[1,2,3,4,5]]).plot(kind = 'bar')
plt.axhline(y=0.9, color='r', linestyle='--')
```

```python
value = pd.DataFrame(ex)
value.columns = ['value']
value['comp'] = [x for x in range(1,n_components+1)]
plt.figure(figsize = (12,6))
plt.title('How many clusters should we choose？')
plt.xlabel('number of components')
plt.ylabel('Explain scale')
plt.bar(  value['comp'],value['value'] )
plt.axhline(y=0.9, color='r', linestyle='--' ,label = '90% explain')
plt.axhline(y=0.8, color='g', linestyle=':' ,label = '80% explain')
print('Let\'s choose 3 components for explaining 90% data!')
```

```python
model = cluster.KMeans(n_clusters=3)
model.fit(data)
print(model.labels_)
```

```python
from fa_kit import FactorAnalysis
import fa_kit.plotting as fa_plotting



# In[3]:
corr_matrix = data.corr(method='pearson')



# In[4]:
from sklearn import preprocessing
data = preprocessing.scale(data)


fa = FactorAnalysis.load_data_samples(
        data,
        preproc_demean=True,
        preproc_scale=True
        )
fa.extract_components()



# In[8]:
fa.find_comps_to_retain(method='top_n',num_keep=3)


# In[9]:
pd.DataFrame(fa.comps["rot"])
fa.rotate_components(method='varimax')
fa_plotting.graph_summary(fa)

import numpy as np
fas = pd.DataFrame(fa.comps["rot"])
data = pd.DataFrame(data)
fa_score = pd.DataFrame(np.dot(data, fas))
```

```python
fas = pd.DataFrame(fa.comps["rot"])
fas.index = list(stock1.iloc[:,:-3])
def perform_component_mean(df):
    total_importance = []
    for col in df.columns:
        balance = ''
        factors = df[col][fas[col] >0.4].index
        for factor in factors:
            balance += factor + ', '
        print('Factor{} focus on {}'.format(str(col + 1) , balance))
perform_component_mean(fas)
```

# 結論

其實以上均只能作為投資的簡單分析，使用更多資料（如近五年）、搭配財務知識，才能使財務報表分析變得比較有意義。<p>
不過我們也可以發現，財務報表不愧為基本面投資必學的工具，利用它至少能檢視一家公司的獲利因素以及變化。<p>
另外綜合因子分析與財務知識，我們可以推測出這三類群組分別代表：

* 紫色群：經營、獲利能力都還好，但是賺錢能力變化大，當作 **投機股**（比如2327國巨）
* 黃色群：替股東賺錢能力變化小且大於零，適合風險承受度小的投資者，可以當作 **定存股**（好比2330台積電）
* 綠色群：好像只有一個XD 經營手段不錯，但獲利影響了EPS，可以當作 **轉機股**（2317 鴻海）
