所以我們檢查一下：

```python
def detect_potential_exception(df , col):
    """
    Input:
        df:DataFrame
        col: Column name , it must be the continuous variable!
    Output:
        Detect result
    """
    confident_value = abs( ( df[col].mode().iloc[0,] - df[col].median())  / (df[col].quantile(0.75) - df[col].quantile(0.25) ))
    confident_value = round(confident_value , 2)
    if confident_value > 0.8:
        print('According to experience rule , Its is dangerous!' , confident_value)
    else:
        print('Safe!' , confident_value)
detect_potential_exception(orders_df, 'days_since_prior_order')
```

我們想得沒錯，基本上除非雙峰分佈，否則出現這麼高的異常值可以推斷為數據經過了粗略地填補。

```python
def make_skew_transform(df , feature):
    """
    To transform high skew data 
    Input:
        df:DataFrame
        feature:The columns of Variable to predict Y
    Output:
        X_rep : DataFrame which process the data with log transform
    
    """
    skew_var_x = {}
    X_rep = df.copy()
    var_x = df[feature]
    for feature in var_x:
        skew_var_x[feature] = abs(X_rep[feature].skew())
        
    skew = pd.Series(skew_var_x).sort_values(ascending = False)
    print(skew)
    
    var_x_ln = skew.index[skew > 1]
    print('針對偏度大於1的進行對數運算')
    
    for var in var_x_ln:
        #針對小於0的我們先確保讓他大於0，平移資料
        if min(X_rep[var]) <= 0:
            X_rep[var] = np.log(X_rep[var] + abs(min(X_rep[var] + 0.01)))
        else:
            X_rep[var] = np.log(X_rep[var])
    return X_rep
```

# 標準化

Kmeans 演算法是一個「無監督學習」的演算法。
就有點像是考試的申論題，沒有標準答案，適合用在很多unlabeled-data 的情境。
而演算法的流程也相當好懂，它會
1. 先從樣本點鐘隨機挑選K個質心，作為初始的集群中心。
2. 指定每個樣本到它最近的「質心」
3. 移動質心到樣本點的中心
4. 重複步驟2、3，直到使用者定義的邊界（可能是可容許誤差，也可能是最大迭代次數）或者群集分配不再改變。

在Kmeans中，我們可以使用歐式距離或者曼哈頓距離來定義資料間的相似性。
那因為是距離，所以必須先確保所有資料的尺度是相同的。


```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
scaler = MinMaxScaler()
data_transform = scaler.fit_transform(tocluster)

sum_of_squared_distance = []
k_clus = range(1,15)

for k in k_clus:
    model = KMeans(n_clusters=k)
    km = model.fit(data_transform)
    sum_of_squared_distance.append(km.inertia_)
plt.plot(k_clus, sum_of_squared_distance, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('We decide to take 4 cluster',fontsize = 20)
plt.show()  
```

我們可以使用Elbow function來查看應該設定K為多少比較好，有拐點的地方就可以當作我們的K值。
在sklearn中，Kmeans的誤差（集群內誤差平方和）就是model的interia_，所以將它一個個畫出來就行囉！
那為什麼是找拐點呢？這點如果用量化的方式，其實我覺得並不是那麼直觀。
基本上可以說這是最適合目前資料的狀態，想像三國演義中，原本動盪不安的東漢末年（雜亂資料）被三個國家劃分，
太多國家會「急速」掀起外患，太少國家又會「急速」引起內亂，而三個國家是比較洽當的。

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
scaler = MinMaxScaler()
data_transform = scaler.fit_transform(tocluster)

sum_of_squared_distance = []
k_clus = range(1,15)

for k in k_clus:
    model = KMeans(n_clusters=k ,init='k-means++')
    km = model.fit(data_transform)
    sum_of_squared_distance.append(km.inertia_)
plt.plot(k_clus, sum_of_squared_distance, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('We decide to take 4 cluster',fontsize = 20)
plt.show()  
```

我們也可以利用輪廓圖來量化集群的品質，可以做為檢驗樣本在集群中，是否「緊密」連結在一起的度量

```python
from sklearn.cluster import KMeans
n_clusters = 4
model = KMeans(n_clusters= n_clusters, random_state=42 ,init='k-means++').fit(tocluster)
centers = model.cluster_centers_
c_preds = model.predict(tocluster)
print(centers)

plt.figure(figsize = (8,8))
colored = [color[k] for k in c_preds]
plt.scatter(tocluster[4] , tocluster[1] , color = colored)
for ci, c in enumerate(centers):
    plt.plot(c[0] , c[1] , 'o' , markersize = 8 , color = color[ci] , alpha = 0.9 ,label = '' + str(ci))
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
```

```python
cust_prod = pd.crosstab(mt['user_id'] , mt['aisle'])
cust_prod['group'] = c_preds
cust_prod.head()
for group in range(n_clusters):
    print('The group' ,str(group) , 'hot product'  , '\n')
    print(cust_prod[cust_prod['group'] == group].sum().sort_values(ascending=  False).head(5))
    print('=' * 20)
    print('\n')
```

```python
for group in range(n_clusters):
    print('The group' ,str(group) , 'hot product'  , '\n')
    print( (cust_prod[cust_prod['group'] == group].sum()/cust_prod.sum()).drop('group').sort_values(ascending=  False).head(5))
    print('=' * 20)
    print('\n')
```

我們可以初步將這四個群集分成：

0 : 酒類愛好者，喜歡紅酒、白酒、香檳，值得一提的是，不曉得跟耳朵、眼睛護理有什麼關係，可能是在意生活品質的白領階級，藉由高品質的促銷或許能夠達到不錯的營收提升。
