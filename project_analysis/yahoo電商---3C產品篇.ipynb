```python
import requests
from bs4 import BeautifulSoup
url= 'https://tw.bid.yahoo.com/tw/%E9%9B%BB%E8%85%A6%E3%80%81%E5%B9%B3%E6%9D%BF%E8%88%87%E5%91%A8%E9%82%8A-23336-category.html?.r=1550661823&hpp=hp_category_23336'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'} 
res = requests.get(url,headers = headers)
content = res.text
```

## 抓銷售與價格區塊

```python
obj = 'BaseGridItem__price___31jkj'
soup = BeautifulSoup(content,'html.parser')
divs = soup.find_all('span' ,obj)
ls=  []
price_list = []
selling_list = []
for div in divs:
    product_info = div.text
    price = product_info.split('/')[0][1:]
    price_list.append(price)
    sell = 0
    if '售出' in product_info:
        sell = product_info.split('/')[1].partition('件')[0].strip()[3:]
    selling_list.append(sell)
title = 'BaseGridItem__title___2HWui'
title_info = soup.find_all('span' , title)
titles=  []
for title in title_info:
    titles.append(title.text)
love = 'FavButton__likeCount___1vsz1'
love = soup.find_all('span',love)
loves = []
for heart in love:
    loves.append(heart.text)
```

```python
import pandas as pd
import numpy as np
col_names = ['Title','sell_amount' , 'price' , 'like_amount']
df = pd.DataFrame([titles ,selling_list , price_list, loves]).T
df.columns = col_names
df.sell_amount = pd.to_numeric(df.sell_amount)
df.head()
```
