```python
from ptt import Board
from requests.exceptions import ReadTimeout
import time
import random 
from wordcloud import WordCloud
```

# ptt 模組是方便抓取ptt文章的lib , random則是隨機產生請求時間

```python
board_name = "soft_job"
font_path = '/Users/Dennis/Downloads/微软正黑体.ttf'
```

```python
titles = ''
pages = Board(board=board_name)
for page in pages:
    if page.isremoved:
        continue
    time.sleep(random.randint(1,3))
    print("正在抓資料中....",page.title)
    try:
        artical = page.read()
    except ReadTimeout:
        pass
    titles += artical.content + artical.title
    
removeword = ['span','class','f3','https','imgur','h1','_blank','href','rel',
              'nofollow','target','cdn','cgi','b4','jpg','hl','b1','f5','f4',
              'goo.gl','f2','email','map','f1','f6','__cf___','data','bbs'
              'html','cf','f0','b2','b3','b5','b6','原文內容','原文連結','作者'
              '標題','時間','看板', '銘言']
for word in removeword:
    titles = titles.replace(word , '')
```

```python
type(titles)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
removed_word = ['之銘言', '引述' , '心得','Soft_Job' , '例如' ,'公告']
for word in removed_word:
    titles = titles.replace(word , '')
wordcloud = WordCloud(font_path=font_path).generate(titles)
plt.figure( figsize=(10,8), facecolor='k' ,frameon=False)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
